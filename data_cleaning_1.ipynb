Bangalore

```python
cd_b = pd.read_excel(r"C:\Users\HP\OneDrive\Desktop\Guvi\capstone_3\bangalore_cars.xlsx")
```

```python
cd_b.drop(columns = ["car_links"], inplace = True,axis=1)
```

```python
def new_car_details(cd_b):
     cd_b["new_car_detail_dict"] = cd_b["new_car_detail"].apply(ast.literal_eval)
     new_car_details_b = pd.json_normalize(cd_b["new_car_detail_dict"])

     return new_car_details_b

new_car_details_b=new_car_details(cd_b)
```

```python
def new_car_overview(cd_b):
    # Convert the 'new_car_overview' column to a dictionary
    cd_b["new_car_overview_dict"] = cd_b["new_car_overview"].apply(ast.literal_eval)
    
    # Normalize the dictionary to flatten it into a DataFrame
    new_car_overview_b = pd.json_normalize(cd_b["new_car_overview_dict"])
    
    # Function to extract key-value pairs from the 'details' list
    def extract_key_values_dict(details):
        if isinstance(details, list):  # Check if details is a list
            return {item.get("key"): item.get("value") for item in details if isinstance(item, dict)}
        else:
            return {}
    
    # Apply the extraction function and normalize the resulting dictionaries
    new_car_overview = new_car_overview_b["top"].apply(extract_key_values_dict)
    new_car_overview_blr = pd.json_normalize(new_car_overview)
   
    
    return new_car_overview_blr
new_car_overview_blr = (new_car_overview(cd_b))
```

```python
def new_car_specs(cd_b):
    
    cd_b["new_car_specs_dict"] = cd_b["new_car_specs"].apply(ast.literal_eval)

    # function to extract key-value pairs from lists of dictionaries
    def extract_key_value_dicts2(details2):
        return {d['key']: d['value'] for d in details2} if isinstance(details2, list) else {}
    
    # Flatten 'new_car_specs_dict' and remove unnecessary columns
    new_car_specs_b = pd.json_normalize(cd_b['new_car_specs_dict']).drop(columns=["commonIcon"], axis=1)
    
    # Normalize the 'top' column to extract key-value pairs
    new_car_specs_top_blr = pd.json_normalize(new_car_specs_b['top'].apply(extract_key_value_dicts2))
    
    # Normalize and extract key-value pairs for the 'data' column (Engine & Transmission, Dimensions & Capacity, Miscellaneous)
    new_car_specs_data = pd.json_normalize(new_car_specs_b['data'])
    
    # Flatten the 'Engine and Transmission' data (first entry in the 'data' column)
    new_car_specs_data_1 = pd.json_normalize(new_car_specs_data[0])
    new_car_specs_list_blr = new_car_specs_data_1["list"].apply(extract_key_value_dicts2)
    new_car_specs_list_blr = pd.json_normalize(new_car_specs_list_blr)

    # Flatten the 'Dimensions & Capacity' data (second entry in the 'data' column)
    new_car_specs_data_2 = pd.json_normalize(new_car_specs_data[1])
    new_car_specs_list_2_blr = new_car_specs_data_2["list"].apply(extract_key_value_dicts2)
    new_car_specs_list_2_blr = pd.json_normalize(new_car_specs_list_2_blr)

    # Flatten the 'Miscellaneous' data (third entry in the 'data' column)
    new_car_specs_data_3 = pd.json_normalize(new_car_specs_data[2])
    new_car_specs_list_3_blr = new_car_specs_data_3["list"].apply(extract_key_value_dicts2)
    new_car_specs_list_3_blr = pd.json_normalize(new_car_specs_list_3_blr)
    
    # Concatenate all the processed data into a final DataFrame
    bangalore_cars_data = pd.concat([
        new_car_details_b, new_car_overview_blr, new_car_specs_top_blr,
        new_car_specs_list_blr, new_car_specs_list_2_blr, new_car_specs_list_3_blr
    ], axis=1)
    
    return bangalore_cars_data

bangalore_cars_data = new_car_specs(cd_b)
```

```python
# add city column to dataframe
bangalore_cars_data["City"] = "Bangalore"
```

```python
# List of dataframes
all_cities_car_data = [bangalore_cars_data,chennai_cars_data,delhi_cars_data,hyderabad_cars_data,jaipur_cars_data,kolkata_cars_data] 

# 1. rename duplicate columns to unique column names
def rename_duplicate_columns(df):
    # find duplicate columns
    cols = pd.Series(df.columns)
    for dup in cols[cols.duplicated()].unique():
        #rename duplicates by appending a suffix '_1','_2',etc.
        count = 0
        for i in range(len(cols)):
            if cols[i] == dup:
                count += 1
                if count > 1:
                    cols[i] = f"{dup}_{count-1}"
    df.columns = cols
    return df

#apply the function to each dataframe to ensure unique column names

all_cities_car_data = [rename_duplicate_columns(df) for df in all_cities_car_data]

# 2. ensure all dataframes have the same columns
# get the set of all unique column names across all dataframes 

all_columns = set()
for df in all_cities_car_data:
    all_columns.update(df.columns)

# add missing columns to ech dataframe with Nan values
for df in all_cities_car_data:
    missing_cols = all_columns - set(df.columns)
    for col in missing_cols:
        df[col] = pd.NA

# 3 . Concatenate Dataframes with outer join 
# Reset index for each Dataframe to avoid index conflicts
all_cities_car_data = [df.reset_index(drop=True) for df in all_cities_car_data]

#concatenate all dataframes along rows (axis=0) with outer join
combined_df = pd.concat(all_cities_car_data, axis = 0, join = 'outer', ignore_index = True)

print(combined_df.shape)
combined_df.head()
```
